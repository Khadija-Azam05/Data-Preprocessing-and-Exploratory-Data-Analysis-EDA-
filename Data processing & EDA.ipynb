```python
import pandas as pd 
df = pd.read_csv("data/Employee.csv")
print("Shape :" , df.shape)
print("Column :\n" , df.columns.tolist(),"\n")
print("Datatype :\n",df.dtypes,"\n")
print("Missing Values :\n", df.isnull().sum(),"\n")
print("Duplicate Values :\n" , df.duplicated().sum(),"\n")
```

```python
df_col = df[["Name","Age","Salary"]]
print(df_col)
```

```python
print("First 10 rows :\n" , df.head(10))
print("Last 5 rows :\n" , df.tail(5))
```

```python
slic_df = df[(df["Age" ].between(30,40)) & ( df["Department"]=="Engineering")]
print("Employees are engineers(30 to 40 years)" ,slic_df)
```

```python
place_df = df[(df["Work_Location" ]=="New York") & ( df["Performance_Score"] > 8.0)]
print( "Employees from new york with 8.0 performance score :" ,place_df)
```

```python
num_col = df.select_dtypes("number")
print("First 5 row of numerical column :\n" ,num_col.head(5))
print("First 5 row of numerical column :\n" ,num_col.tail(5))
```

```python
df["Gender"] = df["Gender"].replace({ "M" : "Male" , "F": "Female","male": "Male", "female": "Female"})
print(df["Gender"])
```

```python
df.loc[df["Age"] < 0, "Age"] = None 
df.loc[df["Salary"] < 0, "Salary"] = None 
df.loc[df["Salary"] > 1_000_000, "Salary"] = None 
print("Age :\n" ,df["Age"])
print("Salary :\n" ,df["Salary"])
```

```python
df["Department"] = df["Department"].str.strip().str.title()  
df["Work_Locaton"] = df["Work_Location"].str.strip().str.title()          

print("Dataset:\n", df.head())
```

```python
import pandas as pd 
df = pd.read_csv("data/Employee.csv")
df_mean = df.copy()
df_mean["Age"] = df_mean["Age"].fillna(df_mean["Age"].mean())
df_mean["Salary"] = df_mean["Salary"].fillna(df_mean["Salary"].mean())

df_median = df.copy()
df_median["Age"] = df_median["Age"].fillna(df_median["Age"].median())
df_median["Salary"] = df_median["Salary"].fillna(df_median["Salary"].median())

print("With Mean:\n", df_mean, "\n")
print("With Median:\n", df_median,"\n")
print(df_mean["Age"] == df_median["Age"])
print(df_mean["Salary"] == df_median["Salary"])
```

```python
df_mode = df.copy()
df_mode["Department"] = df_mode["Department"].fillna(df_mode["Department"].mode()[0])

df_unknown = df.copy()
df_unknown["Department"] = df_unknown["Department"].fillna("Unknown")

print("With Mode:\n", df_mode, "\n")
print("With Unknown:\n", df_unknown)
```

```python
import pandas as pd 
df = pd.read_csv("data/Employee.csv")
num_cols = ["Age", "Salary", "Experience_Years", "Performance_Score"]

for col in num_cols:
    print(f"\nStatistics for {col}:")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
    print("Variance:", df[col].var())
    print("Standard Deviation:", df[col].std())
    print("IQR:", df[col].quantile(0.75) - df[col].quantile(0.25))
    print("Min:", df[col].min())
    print("Max:", df[col].max())
    print("Range:", df[col].max() - df[col].min())

```

```python
import pandas as pd 
df = pd.read_csv("data/Employee.csv")

avg_salary = df.groupby("Department")["Salary"].mean()
print("Average salary per department:\n", avg_salary)
```

```python
avg_perf = df.groupby("Gender")["Performance_Score"].mean()
print("Average performance score per gender:\n", avg_perf)
```

```python
exp_stats = df.groupby("Work_Location")["Experience_Years"].agg(["min", "max", "mean"])
print("Experience Stats per Location:\n", exp_stats)
```

```python
summary = df.groupby("Department")[["Salary", "Experience_Years"]].agg(["mean", "median", "std"])
print("Summary Table (Salary & Experience):\n", summary)
```

```python
pivot_salary = pd.pivot_table(df, values="Salary", index="Department", columns="Gender", aggfunc="mean")
print("Pivot Table : Salary by Department & Gender:\n", pivot_salary)
```

```python
pivot_perf = pd.pivot_table(df, values="Performance_Score", index="Department", columns="Work_Location", aggfunc="mean")
print("Pivot Table - Performance by Department & Location:\n", pivot_perf)
```

```python
import pandas as pd

emp = pd.read_csv("data/Employee.csv")
train = pd.read_csv("data/employee_training_dataset.csv")

emp["Employee_ID"] = "E" + emp["Employee_ID"].astype(str).str.rjust(3, "0")

df = pd.merge(emp, train, on="Employee_ID", how="inner")

print(df.head())
```

```python
df["Salary_per_TrainingHour"] = df["Salary"] / df["Training_Hours"]
print(df[["Employee_ID", "Salary", "Training_Hours", "Salary_per_TrainingHour"]].head())
```

```python
salary_comparison = df.groupby("Certification_Status")["Salary"].mean()
print(salary_comparison)
```
